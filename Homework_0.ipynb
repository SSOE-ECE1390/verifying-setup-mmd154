* Load an (Grayscale) image  (‘Data/Apollo_11_Launch.jpg’)
* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image

```python
import cv2
import os 
import numpy as np 
import matplotlib.pyplot as plt 

#Load and display the image
image=cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'))
plt.imshow(image)

#crop the image 
cropped_region=image[100:700, 400:800]
plt.imshow(cropped_region)

#brighten the image
matrix=np.ones(cropped_region.shape, dtype="uint8") * 75
image_bright=cv2.add(cropped_region,matrix)
plt.imshow(image_bright)

#rotate the image
(h, w) = image_bright.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image_bright, M, (w, h))

plt.imshow(rotated)

#save the image
cv2.imwrite('Data/Apollo_11_Launch_edited.jpg', rotated)
```
